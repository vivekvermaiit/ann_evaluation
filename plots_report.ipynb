```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Paths to your saved figures
img_paths = [
    "sift1m_plain_vs_pq.png",
    "dragon_plain_vs_pq.png",
    "star_plain_vs_pq.png",
]

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for ax, img_path in zip(axes, img_paths):
    img = mpimg.imread(img_path)
    ax.imshow(img)
    ax.axis("off")

plt.tight_layout()
plt.savefig("plots_for_report/kannolo_plain_vs_pq.png", dpi=300, bbox_inches="tight")
plt.show()

#First plot impact of PQ
```
