```python
import pandas as pd
import matplotlib.pyplot as plt

# Load TSV files
plain = pd.read_csv("sift_plain_report.tsv", sep="\t")
pq = pd.read_csv("sift_pq_report.tsv", sep="\t")

# Convert accuracy from percentage to decimal
plain["accuracy"] = plain["Accuracy"] / 100.0
pq["accuracy"] = pq["Accuracy"] / 100.0

# Convert query time (microseconds) to QPS
plain["qps"] = 1e6 / plain["Query Time (microsecs)"]
pq["qps"] = 1e6 / pq["Query Time (microsecs)"]

# Plot
plt.figure(figsize=(7, 5))

plt.plot(
    plain["accuracy"],
    plain["qps"],
    marker="o",
    linestyle="-",
    color="red",
    label="kANNolo plain",
)

plt.plot(
    pq["accuracy"],
    pq["qps"],
    marker="+",
    linestyle="-",
    color="blue",
    label="kANNolo PQ",
)

# Log-scale Y axis
plt.yscale("log")
plt.yticks(
    [ 1e3, 1e4],
    [r"$10^3$", r"$10^4$"],
)

# X axis ticks every 0.05
plt.xticks([i / 100 for i in range(70, 101, 5)])

plt.xlabel("Accuracy@10")
plt.ylabel("Queries per second (1/s)")
plt.title("SIFT1M (1M, 128d)")

plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.tight_layout()
plt.savefig("sift1m_plain_vs_pq.png", dpi=300, bbox_inches="tight")

plt.show()
```
